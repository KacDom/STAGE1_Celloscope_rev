```python
import pandas as pd
import numpy as np
import os
import glob
import heapq
dict_res_all = {}
```

```python
ass_res_dir_path = "/Users/kd/Desktop/cell2location/dense_tutorial_params"
setup_type = ass_res_dir_path.split('/')[-1]
res_dirs = sorted([path for path in os.listdir(ass_res_dir_path) if not path.startswith(".")])
```

```python
for path in res_dirs:
    os.chdir(ass_res_dir_path + "/" + path)
    df_true = pd.read_csv("TRUE_h.csv", index_col=0)
    df_est = pd.read_csv(path + ".csv", index_col=0)
    df_true.columns = df_est.columns
    df_true.index = df_est.index
    dict_temp = (np.mean(np.abs(df_true - df_est)))
    dict_temp = dict_temp.add_suffix("_mean_err")
    dict_temp = dict_temp.to_dict()
    dict_temp['total_mean_err'] = np.mean(np.mean(np.abs(df_true - df_est)))
    dict_res_all[setup_type + '_' + path] = dict_temp
```

```python
pd.DataFrame(dict_res_all).transpose() #.to_csv(ass_res_dir_path + '/' + "BayesPrism_gene_num_test.csv")
```
